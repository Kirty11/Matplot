```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas=["Marketing","Sales","Development","HR","Customer Support"]
budget=[2.5,4,10,1,6]
```

```python
plt.pie(budget,labels=areas,autopct="%0.2f%%")
plt.show()
```

# Bar Graph

```python
year=[2013,2014,2015,2016,2017,2018]
Prices=[45600,78900,43200,123499,78999,90000]
plt.bar(year,Prices,color="Green")
plt.xlabel("Year")
plt.ylabel("Prices(In INR)")
plt.title("Year Vs Price")
plt.show()
```

# Scatter Graph

```python
a=[2,3,4,5,1,8,9]
b=[100,200,300,400,500,600,700]
plt.scatter(a,b)
plt.show()
```

# 2-D Graph

```python
plt.plot(a,b,color="red",marker='*',markersize=8)
plt.show()
```
